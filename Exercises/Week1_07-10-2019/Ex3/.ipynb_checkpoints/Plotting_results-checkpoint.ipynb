```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""
RETRIEVING RESULTS FROM THE FILE PRINTED BY THE TERMINAL
N.B.: this procedure is specific for this case (this easy algoritm simply look for a "." in the line and from
    this it infers there is a number)
"""

# file name input
in_file_name = "my_res.txt"

# opening files
infile = open(in_file_name, "r")
temporary_list = []
for line in infile:
    app = str(line)
    if "." in app:
        temporary_list.append(float(app))
    
infile.close()

temporary_list = np.array(temporary_list)

opt0 = temporary_list[0:3]
opt1 = temporary_list[3:6]
opt2 = temporary_list[6:9]
opt3 = temporary_list[9:12]

print(opt0)
print(opt1)
print(opt2)
print(opt3)
```

```python
plt.figure(figsize=(15, 10))
plt.title("Diverse ottimizzazioni; dim=500", fontsize=20)
plt.plot(opt0, label="No opt")
plt.plot(opt1, label="Opt1")
plt.plot(opt2, label="Opt2")
plt.plot(opt3, label="Opt3")

plt.xticks(ticks=[0, 1, 2], labels=["Manual", "Transposed", "Matmul"], fontsize=15)

plt.ylabel("Times [s]", fontsize=15)
plt.legend()
plt.show()
```
